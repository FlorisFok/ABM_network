```python
import model
import numpy as np
import pickle

cons = ({'type' : 'ineq', 'fun' : lambda setting: 5 - setting[0]},  # delta[0, 5]
        {'type' : 'ineq', 'fun' : lambda setting : setting[0] - 10 ** (-6)},
        {'type' : 'ineq', 'fun' : lambda setting : 1 - setting[1]},  # gamma[0, 1]
        {'type' : 'ineq', 'fun' : lambda setting : setting[1] - 10 ** (-6)},
        {'type' : 'ineq', 'fun' : lambda setting : 5 - setting[2]},  # C[0, 5]
        {'type' : 'ineq', 'fun' : lambda setting : setting[2] - 10 ** (-6)},
        {'type' : 'ineq', 'fun' : lambda setting : 5 - setting[3]},  # b123[0, 5]
        {'type' : 'ineq', 'fun' : lambda setting : setting[3] - 10 ** (-6)},
        {'type' : 'ineq', 'fun' : lambda setting : 5 - setting[4]},  # b123[0, 5]
        {'type' : 'ineq', 'fun' : lambda setting : setting[4] - 10 ** (-6)},
        {'type' : 'ineq', 'fun' : lambda setting : 5 - setting[6]},  # b123[0, 5]
        {'type' : 'ineq', 'fun' : lambda setting : setting[6] - 10 ** (-6)},
        {'type' : 'ineq', 'fun' : lambda setting : 5 - setting[3]},  # sigma[0, 5]
        {'type' : 'ineq', 'fun' : lambda setting : setting[3] - 10 ** (-6)}
        )

def check(settings, cons):
    for i in cons:
        if not ((i['fun'](settings) >= 0)):
            break
    else:
        return True
    return False

settings =  [0.05, 0.65, 0.175, 0.035, 0.09999999999999998, 0.09999999999999998, 0.19999999999999998]
settings =  [0.05, 0.65, 0.175, 0.035, 0.5, 0.5, 0.5]

var = {'av_degree':0, 'clustering_coefficient':2, 'RHO2':1,
                   'segreg_ind0':3, 'segreg_ind1':4, 'segreg_ind2':5}

_string = ['av_degree', 'RHO2', 'clustering_coefficient', 'segreg_ind0', 'segreg_ind1', 'segreg_ind2']

to_change = ['segreg_ind0', 'segreg_ind1', 'segreg_ind2']

def avg(l, reruns):
    return [sum(l[i*reruns:(i+1)*reruns])/reruns for i in range(int(len(l)/reruns))]

def true_data_make(max_len):
    g_matrix = pickle.load(open(r"C:\Users\FlorisFok\Downloads\g_list.pkl", 'rb'))
    big_x = pickle.load(open(r"C:\Users\FlorisFok\Downloads\x_list.pkl", 'rb'))
    
    true_data = {'av_degree':[], 'clustering_coefficient':[], 'RHO2':[],
                 'segreg_ind0':[], 'segreg_ind1':[], 'segreg_ind2':[]}
    
    for num, X in enumerate(big_x):
        g = g_matrix[num]
        if len(X) > max_len:
            continue
        output2 = model.analyse_network(g, X)
        for s, o in zip(_string, output2):
            true_data[s].append(o)
    
#     for s in _string:
#         true_data[s] = avg(true_data[s], len(true_data[s]))
    
    return true_data

def scoring(plot_data, true_data):
    best = 0
    for t in _string:
        for tv, v in zip(true_data[t], plot_data[t]):
            best += abs(tv - v)
    
    return best/(len(true_data[_string[0]]))


max_len = 50
stepsize = 0.1
change = 6
true_data = true_data_make(max_len)

derivitive = np.random.choice([-1,1], size=len(settings))
plot_data = model.main(settings, max_len)
base = scoring(plot_data, true_data)
progress = [base]
pos_progress = [base]

best = [base, settings]

            
for i in range(100):
    print(base, settings)
    
    if best[0] > base:
        best[0] = base
        best[1] = settings
    
    change = [sett for sett in settings]
    for s in to_change:
        n = var[s]
        
        positive = True
        change[n] = settings[n] + (stepsize * derivitive[n])
        
        if not check(change, cons):
            change[n] = settings[n] - (stepsize * derivitive[n] * -2)
            derivitive[n] *= -1
            positive = False

        if check(change, cons):
            plot_data = model.main(change, max_len)
            new = scoring(plot_data, true_data)
            progress.append(new)
            if (new - base) < 0:
                settings[n] = round(change[n], 6)
                base = new
                pos_progress.append(base)
            else:
                derivitive[n] *= -1 
        
        if progress[-change:].count(progress[-1]) == change:
            stepsize = stepsize/2

print()
print(best)
```

```python
import model
import numpy as np
import matplotlib.pyplot as plt

settings =  [0.05000000000000004, 0.6499999999999999, 0.175, 0.035, 0.09999999999999998, 0.09999999999999998, 0.19999999999999998]


var = {'av_degree':0, 'clustering_coefficient':2, 'RHO2':1,
                   'segreg_ind0':3, 'segreg_ind1':4, 'segreg_ind2':5}

_string = ['av_degree', 'RHO2', 'clustering_coefficient', 'segreg_ind2', 'segreg_ind1', 'segreg_ind0']

all_data =  {'av_degree':{}, 'clustering_coefficient':{}, 'RHO2':{},
                 'segreg_ind0':{}, 'segreg_ind1':{}, 'segreg_ind2':{}}

MAX = 200

for var_name in _string:
    print(var_name)
    
    for i in np.arange(0, 2, 0.1):
        settings2 = settings
        settings2[var[var_name]] = i
        data = model.main(settings2, MAX)
        all_data[var_name][i] = data[var_name]
        
```
