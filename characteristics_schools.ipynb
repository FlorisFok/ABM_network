```python
import pickle
import pandas as pd
import numpy as np
```

```python
# Import the data from pickle files
g_list = pickle.load(open("g_list.pkl", "rb"))
X_list = pickle.load(open("X_list.pkl", "rb"))
```

```python
# make a dataframe of the proportion of students in different cathegories

school_matrices = []

for school_pd in X_list:
    # determine the relative frequency of occurences of characteristics
    sex = school_pd['sex'].value_counts(normalize=True)
    race = school_pd['race'].value_counts(normalize=True)
    grade = school_pd['grade'].value_counts(normalize=True)
    school = school_pd['school'].value_counts(normalize=True)
    
    # create empty datamatrices
    students_mat = np.zeros((1))
    sex_mat = np.zeros((3))
    race_mat = np.zeros((6))
    grade_mat = np.zeros((13))
    school_mat = np.zeros((3))
    
    # adapt the datamatrix values to the relative frequencies
    students_mat[0] = school_pd['sex'].count()
    for n_sex in sex.index:
        sex_mat[n_sex] = sex[n_sex]
    for n_race in race.index:
        race_mat[n_race] = race[n_race]
    for n_grade in grade.index:
        grade_mat[n_grade] = grade[n_grade]
    for n_school in school.index:
        school_mat[n_school] = school[n_school]
        
    # combine all relevant values into a single array for the school
    school_mat = np.concatenate((students_mat,sex_mat[1:],race_mat[1:],grade_mat[7:],school_mat[1:]))
    
    school_matrices.append(school_mat)
    
# convert the list of schools to a matrix with schools in rows
school_matrices = np.array(school_matrices)

# convert the matrix to a pandas dataframe with titles of characteristics
school_characteristics = pd.DataFrame(data=school_matrices)
school_characteristics.columns = ['n_students','male','female','white','black','hispanic','asian','mixed / other','grade 7','grade 8','grade 9','grade 10','grade 11','grade 12','school 1','school 2']
```
